## Coherence scores
    
    We used the UMass coherence score, which calculates how often two words, w_{i} and w_{j} appear together in the corpus. The greater the number, the better is the coherence score.
    



```r
library(tidyverse)
```

```r
data <- read_csv("../results/fine-scale/all-countries/calculated_coherence.csv") %>%
    select(-1)
```

```r
head(data)
```

```r
options(repr.plot.width=20, repr.plot.height=10)

data %>%
    # pivot_longer(-Topics, names_to = "Metric", values_to = "Value") %>%
    # # group_by(Metric) %>%
    # # mutate(Value = (Value) / max(Value)) %>%
    ggplot(aes(x=Topics,y=umass))+
        geom_line()+
        theme_bw(base_size=20)
```
